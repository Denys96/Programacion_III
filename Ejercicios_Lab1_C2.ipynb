Utilizando las librerías obtenga la utilización de tarjetas de crédito: por sexo (cuenta) .

```python
import pandas as pd
import matplotlib.pyplot as plt

archivo = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",", encoding="utf-8")

proceso = pd.crosstab(index=archivo["Gender"],columns="frecuencia")
print(proceso)

tarjetasexo= archivo.groupby(["Gender","Card Type"]).agg({"sum"})

plt.pie(proceso, labels=proceso.index, autopct="%1.1f%%")
plt.xlabel(f"Utilización de tarjeta de credito por sexo, {proceso}")
plt.show()
proceso.to_excel("/content/Utilizacion_de_tarjeta_por_sexo.xlsx")
```

TOTAL EN DOLARES UTILIZADOS DIFERENCIANDO TARJETAS DE CREDITO POR CIUDAD

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

archivo = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",", encoding="utf-8")

proceso = pd.crosstab(index=archivo["City"],columns="frecuencia")
print(proceso.head(10))

tarjetasciudad= archivo.groupby(["City","Card Type"])["Amount"].agg({"sum",})

plt.pie(tarjetasciudad.head(10),autopct="%1.1f%%", shadow= True,radius=3 )
plt.legend(loc=2, fontsize=9,labels=tarjetasciudad.index)
plt.title("Total de Dls por ciudades y tipo de tarjeta, Primeras 10 ciudades")
plt.axis("equal")
plt.xlim(4,3)
plt.show()

tarjetasciudad.to_excel("/content/Utilizacion_de_tarjeta_por_ciuda.xlsx")
```

Total utilizado por tipo de gasto

```python
import pandas as pd
import matplotlib.pyplot as plt

archivo = pd.read_csv("/content/trx_tarjetas_credito.csv", sep=",", encoding="utf-8")

proceso = pd.crosstab(index=archivo["Exp Type"],columns="total")
print(proceso)

gastostotal= archivo.groupby(["Exp Type"])["Amount"].agg({"sum"})

plt.pie(proceso.head(25),autopct="%1.1f%%", shadow= True,radius=5 )
plt.legend(loc=2, fontsize=10,labels=proceso.index)
plt.title("Total de tipo de gasto")
plt.axis("equal")
plt.xlim(6,2)
plt.show()

gastostotal.to_excel("/content/Utilizacion_de_tarjeta_por_gasto.xlsx")
```
